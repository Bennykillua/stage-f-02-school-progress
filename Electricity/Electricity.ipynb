## Access to electricity by Indian Schools in different school categories (2013-2016).

### Import relevant libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Load data

```python
elect = pd.read_csv('../data/percentage-of-schools-with-electricity-2013-2016.csv')
elect.head()
```

### Explore data

```python
elect.info()
```

 #### Find how many indian states are in the data and the unique values of the 'State_UT' column.

```python
elect['State_UT'].nunique()
```

```python
elect['State_UT'].unique()
```

#### Find the academic years of the Indian Schools that the data was collected i.e. The unique values in the 'year' column.

```python
elect['year'].unique()
```

#### Find the different school categories and levels provided in the data used to compute the access to electricity .

```python
elect_melted = pd.melt(elect, id_vars=['State_UT', 'year'], var_name='Categories', value_name='Values')
elect_melted.head()
```

```python
elect_melted['Categories'].unique()
```

```python
school_levels_only = elect[['State_UT','year','Primary_Only', 'U_Primary_Only', 'Sec_Only', 'HrSec_Only', 'All Schools']]
school_levels_only.head()
```

##### This implies that there are 37 unique values in 'State_UT' column of which 36 of these values are India States. There are also three unique values for the 'year'column i.e. there are three academic years for the analysis. The unique School levels are Primary, Upper primary, Secondary and Higher Secondary Levels.

### Perform Analysis and Visualization of each academic year using 'All Schools' level

```python
plt.style.use('seaborn')
#sns.set_theme(style="whitegrid")
plt.rc('figure', figsize=(10, 8))  # to make all the graph the same size
```

```python
size = (24, 12)
```

#### Which Indian States have 24/7 access to electricity in all school levels ?

```python
all_schools = school_levels_only.groupby('State_UT')['All Schools'].mean()
all_schools.plot(kind='bar')
```

Some Indian states have less than 50% access to electricity for all school categories namely Arunachal Pradesh, Assam, Bihar, Jammu and Kashmir, Jharkhand, Madhya Pradesh, Manipur, Meghalaya, Odisha and Tripura. Most of these states mentioned are underdeveloped and rural areas in India. 
States with 24/7 access to electricity in all school categories include Daman & Diu, Chandigarh, Delhi, Lakshadweep and Puducherry.

#### Has there been improvement in access to electricity from previous academic year to the current academic year in all school levels in India States?

### A) 2013-14_All Schools

```python
year_2013 = school_levels_only[school_levels_only['year']=='2013-14']
year_2013.head()
```

```python
year_2013_sorted = year_2013.set_index('State_UT').sort_values('All Schools', ascending=False)
year_2013_sorted.head()
fig, ax = plt.subplots(figsize=size)
sns.barplot(year_2013_sorted.index, year_2013_sorted['All Schools'],  palette="Blues_d")
plt.xlabel('Indian States')
plt.ylabel('Percentage of Electricity')
plt.xticks(rotation=90, fontsize=15)
plt.title('Access to Electricity in Indian Schools for the 2013-14 academic year', fontsize=15)
plt.show()
```

### 2014-15_All Schools

```python
year_2014 = school_levels_only[school_levels_only['year']=='2014-15']
year_2014.head()
```

```python
year_2014_sorted = year_2014.set_index('State_UT').sort_values('All Schools', ascending=False)
year_2014_sorted.head()
fig, ax = plt.subplots(figsize=size)
sns.barplot(year_2014_sorted.index, year_2014_sorted['All Schools'], palette="Blues_d")
plt.xlabel('Indian States')
plt.ylabel('Percentage of Electricity')
plt.xticks(rotation=90, fontsize=15)
plt.title('Access to Electricity in Indian Schools for the 2014-15 academic year', fontsize=15)
plt.show()
```

### 2015-16_All Schools

```python
year_2015 = school_levels_only[school_levels_only['year']=='2015-16']
year_2015.head()
```

```python
year_2015_sorted = year_2015.set_index('State_UT').sort_values('All Schools', ascending=False)
year_2015_sorted.head()
fig, ax = plt.subplots(figsize=size)
sns.barplot(year_2015_sorted.index, year_2015_sorted['All Schools'], palette="Blues_d")
plt.xlabel('Indian States')
plt.ylabel('Percentage of Electricity')
plt.xticks(rotation=90, fontsize=15)
plt.title('Access to Electricity in Indian Schools for the 2015-16 academic year', fontsize=15)
plt.show()
```

There is no serious improvement in the access to electricity in all the school levels in Indian States especially the rural areas from 2013-14 academic session to 2015-16 academic session. States like Assam, Bihar and Jharkhand did not make much improvement in access to electricity.

#### Is there improvement in access to electricity in each school level from one academic session to another academic session in all India States?

```python
all_india = school_levels_only[school_levels_only['State_UT']=='All India']
all_india.head()
all_india_df = pd.melt(all_india, id_vars=['State_UT','year'], var_name='School_Levels', value_name='Values')
all_india_df.head()
all_india_df = all_india_df.groupby(['year', 'School_Levels'], as_index=False).sum()
all_india_df.head()
```

```python
fig, ax = plt.subplots(figsize=size)
bar = sns.barplot(x="year",y='Values',hue='School_Levels',data=all_india_df ,palette = 'deep',edgecolor='black',ax=ax)
bar.set(ylim=(0, 125))
# highlighting the title
ax.set_title('Percentage of Schools with Electricity in India in three academic years',size = 16 , pad = 16, color='white',bbox=dict(facecolor='black', alpha=1.0))

ax.set_ylabel('Percentage',size=13)
ax.set_xlabel('Year',size=13)

bar.legend(fancybox=True, framealpha=1, borderpad=0.5)
```

The access to electricity in each of the school categories (i.e. Primary, Upper Primary, Secondary and Higher Secondary) improved from one academic session to another academic session.

#### Which India States had 24/7 access to elecricity in Primary School Level across all the academic years?

```python
schools_melted = pd.melt(school_levels_only, id_vars=['State_UT','year'], var_name='School_Levels', value_name='Values')
#schools_melted
primary_schools = schools_melted[schools_melted['School_Levels']=='Primary_Only']
primary_schools = primary_schools.groupby('State_UT', as_index=False).mean()
primary_schools.set_index('State_UT').sort_values('Values').plot(kind='bar')
```

The Indian States that have 24/7 access to electricity in Primary Schools across the three academic sessions are Punjab, Puducherry, Chandigarh, Delhi, Lakshadweep and Daman & Diu.

#### Which India States have 24/7 access to electricity in Upper Primary Schools across the three academic years?

```python
upper_primary_schools = schools_melted[schools_melted['School_Levels']=='U_Primary_Only']
upper_primary_schools = upper_primary_schools.groupby('State_UT', as_index=False).mean()
upper_primary_schools.set_index('State_UT').sort_values('Values').plot(kind='bar')
```

Indian States that have 24/7 access to electricity in Upper Primary Schools across the three academic years include Punjab, Gujarat, Goa, Sikkim, Dadra & Nagar Haveli, Lakshadweep, Daman & Diu, Delhi and Andhra Pradesh.

#### Which India States have 24/7 access to electricity in Secondary Schools across the three academic years?

```python
secondary_schools = schools_melted[schools_melted['School_Levels']=='Sec_Only']
secondary_schools = secondary_schools.groupby('State_UT', as_index=False).mean()
secondary_schools.set_index('State_UT').sort_values('Values').plot(kind='bar')
```

Indian States with 24/7 access to electricity in Secondary Schools across the three academic years are Himachal Pradesh, Punjab, Daman & Diu, Tamil Nadu and Dadra & Nagar Haveli. Some states with 24/7 access to electricity in  Upper Primary or Primary Levels are having no access to electricity in Secondary Schools e.g Lakshadweepp, Puducherry, Delhi and Chandigarh.

#### Which India States have 24/7 access to electricity in Higher Secondary Schools across the three academic years?

```python
higher_secondary_schools = schools_melted[schools_melted['School_Levels']=='HrSec_Only']
higher_secondary_schools = higher_secondary_schools.groupby('State_UT', as_index=False).mean()
higher_secondary_schools.set_index('State_UT').sort_values('Values').plot(kind='bar')
```

Some India States with 24/7 access to electricity in Higher Secondary Schools across the three academic years are Himachal Pradesh, Puducherry, Goa, Tamil Nadu, Daman & Diu etc. 

##### Summary:
-  There have been improvement in access to electricity for each School Level in all Indian States from one academic session to another.
- Rural areas have not shown much improvement in access to electricity across the academic sessions.
- Some India States like Chandigarh and Lakshadweep with 100% access (i.e. 24/7) to electricity in their Primary and Upper Primary Schools have no access to electricity in the Secondary Schools and Higher Secondary Schools.
