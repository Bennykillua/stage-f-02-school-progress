```python
# Importing the necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading the toilet facilities datasets
boys_toilet = pd.read_csv('../input/indian-school-education-statistics/schools-with-boys-toilet-2013-2016.csv')
girls_toilet = pd.read_csv('../input/indian-school-education-statistics/schools-with-girls-toilet-2013-2016.csv')
```

## Boys Toilet Facilities

```python
boys_toilet.head()
```

```python
boys_toilet.isnull().sum()
```

```python
excl_year = boys_toilet.drop('year', axis = 1)
excl_year = excl_year.groupby('State_UT').mean()
```

```python
# Mean of male toilet facilities in each state from 2013-2016
excl_year.head()
```

```python
#To see the total male toilet facilities distribution amongst each school from 2013-2016
df = pd.DataFrame([(colname, excl_year[colname][i]) for i in range(len(excl_year)) for colname in excl_year.columns], 
                 columns=['col', 'values'])
plt.figure(figsize = (30, 15))
sns.swarmplot(x = 'col', y='values', data=df)
```

```python
#To see the total male toilet facilities distribution amongst each state from 2013-2016
excl_year.plot.bar(figsize = (30, 15), stacked=True)
plt.legend(loc="upper center", bbox_to_anchor=(0.5, 1.15), ncol=2)
```

The graph shows that the states of Kamataka and Mizoram have the highest and lowest number of male toilet facilities respectively, as comapared to other states.

```python
#To see the total male toilet facilities distribution for each year
each_year = boys_toilet.groupby('year').mean()
each_year.plot.bar(figsize = (10, 10), stacked=True)
plt.legend(loc="upper center", bbox_to_anchor=(0.5, 1.15), ncol=2)
```

The total number of boys toilet facilities in 2015-16 is the highest compared to previous years. The graph shows an annual growth of male toilet facilities

## Girls Toilet Facilities

```python
girls_toilet.head()
```

```python
girls_toilet.isnull().sum()
```

```python
exc_year = girls_toilet.drop('year', axis = 1)
exc_year = exc_year.groupby('State_UT').mean()
```

```python
# Mean of female toilet facilities in each state from 2013-2016
exc_year.head()
```

```python
#To see the total female toilet facilities distribution amongst each school from 2013-2016
df = pd.DataFrame([(colname, exc_year[colname][i]) for i in range(len(exc_year)) for colname in exc_year.columns], 
                 columns=['col', 'values'])
plt.figure(figsize = (30, 15))
sns.stripplot(x = 'col', y='values', data=df)
```

```python
#To see the total female toilet facilities distribution amongst each state from 2013-2016
exc_year.plot.bar(figsize = (30, 15), stacked=True)
plt.legend(loc="upper center", bbox_to_anchor=(0.5, 1.15), ncol=2)
```

Just like the male category, the state of Kamataka has the highest total number of female toilet facilities while Mizoram has the lowest total number of female toilet facilities from 2013-16, as compared to other states

```python
#To see the total female toilet facilities distribution for each year
each_year = girls_toilet.groupby('year').mean()
each_year.plot.bar(figsize = (10, 10), stacked=True)
plt.legend(loc="upper center", bbox_to_anchor=(0.5, 1.15), ncol=2)
```

The total number of girls toilet facilities in 2015-16 is the highest, although, almost as equal as the year 2014-15. The graph also shows an annual growth of female toilet facilities
