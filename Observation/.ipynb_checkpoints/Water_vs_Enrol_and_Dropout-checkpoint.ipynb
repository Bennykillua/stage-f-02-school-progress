<b>Merging and Comparing the datasets to find out if water is a factor in  gross enrollment and dropout ratio</b>

```python
#Function to compare State columns to ensure consistency across the three datasets during merging
def compare_columns(f_cols,s_cols,f_name,s_name):
    f_val = []
    s_val = []
    
    for vals in f_cols:
        f_val.append(vals)
    
    for valss in s_cols:
        s_val.append(valss)
        
        
    print("Values in {} only".format(f_name))
    for n in f_val:
        if (n not in s_val):
            print(n)
    print("------------------------------")
    
    print("Values in {} only".format(s_name))
    for k in s_val:
        if (k not in f_val):
            print(k)
```

```python
compare_columns(water["State_UT"],enrol["State_UT"],"water","enrollment")
```

```python
compare_columns(water["State_UT"],dropout["State_UT"],"water","dropout")
```

```python
dropout["State_UT"].replace({"A & N Islands":"Andaman & Nicobar Islands","Jammu & Kashmir":"Jammu And Kashmir"},inplace=True)
```

```python
compare_columns(water["State_UT"],dropout["State_UT"],"water","dropout")
```

```python
#Adding a sufix to each dataframe to be able to differentate when merged

enrol.columns=enrol.columns.map(lambda x: x +"_enrol" if x !="State_UT" and x !="Year" else x)

water.columns = water.columns.map(lambda x: x+"_water" if x != "State_UT" and x!= "Year" else x )

dropout.columns = dropout.columns.map(lambda x: x+"_drop" if x!= "State_UT" and x!="Year" else x)
```

```python
#Splitting into the individual years
water13 = water[water["Year"]=="2013-14"]
water14 = water[water["Year"]=="2014-15"]
water15 = water[water["Year"]=="2015-16"]

enrol13 = enrol[enrol["Year"]=="2013-14"]
enrol14 = enrol[enrol["Year"]=="2014-15"]
enrol15 = enrol[enrol["Year"]=="2015-16"]

#Dropout starts from 2012-13 and doesn't have 2015-16
dropout13 = dropout[dropout["Year"]=="2013-14"]
dropout14 = dropout[dropout["Year"]=="2014-15"]
```

```python
compare_columns(water15["State_UT"],water13["State_UT"],"'15","'13")
```

```python
compare_columns(enrol15["State_UT"],enrol13["State_UT"],"'15","'13")
```

```python
compare_columns(dropout14["State_UT"],dropout13["State_UT"],"'15","'13")
```

```python
#Merging of the datasets

we13 = pd.merge(water13,enrol13,on="State_UT")
total13 = pd.merge(we13,dropout13,on="State_UT")

we14 = pd.merge(water14,enrol14,on="State_UT")
total14 = pd.merge(we14,dropout14,on="State_UT")

total15 = pd.merge(water15,enrol15,on="State_UT")
```
